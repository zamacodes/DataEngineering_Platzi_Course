```python
import bs4
soup = bs4.BeautifulSoup(response.text, 'html.parser')

#print(soup.title.text)
#print(soup.select('meta[name=description]')[0]['content'])
categories_links = soup.select('.Category')
categories = [category['href'] for category in categories_links ]

for category in categories:
    print(category)
```

```python
course_links = soup.select('.HomeCategories-category a')
courses = [course['href'] for course in courses_links ]

for course in courses:
    print(courses)
```
